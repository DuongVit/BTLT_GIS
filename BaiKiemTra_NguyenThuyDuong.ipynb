#### `Câu 18: Sử dụng bộ thư viện pandas, geopandas, matplotlib.pyplot và dữ liệu lấy từ thư mục 'data/berlin-districts.geojson'. Hiển thị lược đồ cho cột "district_group":`

#### `Câu 20: Sử dụng thư viện pandas và geopandas, khai báo dữ liệu rivers và countries: Sử dụng hàm geopandas.read_file, file zip có sẵn trong thư mục '/data/ne_50m_rivers_lake_centerlines.zip' và '/data/ne_110m_admin_0_countries.zip'`
- Trích xuất dòng sông Amazon ở Nam Mỹ
- Truy vấn các quốc gia mà dòng sông chảy qua

```python
%matplotlib inline

import pandas as pd
import geopandas

#pd.options.display.max_rows = 20
```

```python
rivers = geopandas.read_file("zip://./data/ne_50m_rivers_lake_centerlines.zip")
countries = geopandas.read_file("zip://./data/ne_110m_admin_0_countries.zip")
```

```python
amazon = rivers[rivers['name'] == 'Amazonas'].geometry.squeeze()
```

```python
countries[countries.crosses(amazon)]  # or .intersects
```
