```python
%matplotlib inline
import pandas as pd
import geopandas as gpd
```

```python
bien = gpd.read_file("data/Bai1_Nhap va Trinh bay DL/Bien.shp")
duong = gpd.read_file("data/Bai1_Nhap va Trinh bay DL/Duong.shp")
huyen = gpd.read_file("data/Bai1_Nhap va Trinh bay DL/huyen.shp")
song = gpd.read_file("data/Bai1_Nhap va Trinh bay DL/Song.shp")
tinh = gpd.read_file("data/Bai1_Nhap va Trinh bay DL/Tinh.shp")
UBND_huyen = gpd.read_file("data/Bai1_Nhap va Trinh bay DL/UBND_huyen.shp")
UBND_tinh = gpd.read_file("data/Bai1_Nhap va Trinh bay DL/UBND_tinh.shp")
xa = gpd.read_file("data/Bai1_Nhap va Trinh bay DL/xa.shp")
```

```python
ax = tinh.plot(edgecolor='k', facecolor='none', figsize=(25, 20),legend=True)
huyen.plot(ax=ax, color = 'black')
xa.plot(ax=ax, color = 'white')
song.plot(ax=ax, color = 'blue')
bien.plot(ax=ax)
duong.plot(ax=ax, color = 'brown')
UBND_tinh.plot(ax=ax, color = 'yellow')
UBND_huyen.plot(ax=ax, color = 'red')
ax.set(xlim=(175000, 1200000), ylim=(900000, 2650000))
```
